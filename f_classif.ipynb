```python
import numpy as np
import pandas as pd
df = pd.read_csv("D:\\feature_selection\\cardio.csv")
df
```

```python
x = df.drop('cardio',axis=1)
x
```

```python
x = df.drop('id',axis=1)
x
```

```python
y = df.cardio
y
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
FEATURES = SelectKBest(score_func = f_classif)
```

```python

FEATURES.fit(x,y)
```

```python
FEATURES.scores_
```

```python
score_col = pd.DataFrame(data = FEATURES.scores_,columns = ['score'])
```

```python
score_col
```

```python
name_col = pd.DataFrame(data = x.columns,columns = ['Features'])
```

```python
name_col
```

```python
Features_score = pd.concat([score_col,name_col],axis=1)
Features_score
```

```python
Features_score.nlargest(8,'score')
```
